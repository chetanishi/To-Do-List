```python
import os
import shutil


os.makedirs('sample_dir', exist_ok=True)

# Create a sample file inside 'sample_dir'
file_path = 'sample_dir/sample_file.txt'
with open(file_path, 'w') as file:
    file.write("This is a sample file.")

# Display the current directory and file structure
print("Current directory structure:")
print(os.listdir('sample_dir'))

# 1. Renaming a file
a = 'sample_dir/renamed_file.txt'
os.rename(file_path, a)
print(f"\nFile renamed to: {renamed_file}")

# 2. Moving a file
new_location = 'sample_dir/moved_file.txt'
shutil.move(renamed_file, new_location)
print(f"\nFile moved to: {new_location}")

# 3. Copying a file
copied_file = 'sample_dir/copied_file.txt'
shutil.copy(new_location, copied_file)
print(f"\nFile copied to: {copied_file}")

# 4. Removing a file
os.remove(copied_file)
print(f"\nFile removed: {copied_file}")

# 5. Removing an empty directory
#os.rmdir('sample_dir')
#print(f"\nDirectory removed: {sample_dir}")
```
