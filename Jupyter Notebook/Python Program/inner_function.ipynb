```python
def outer():
    def inner(grades):
        return sum(grades)/len(grades)

    grades=[78,76,90,98,77]
    result=inner(grades)
    print("The grades is")
    print(f"Grades:{grades}")
    print(f"Avarage grade is:{result:.2f}")

    if result>90:
        print("+A")

    elif result>80:
        print("A")

    elif result>70:
        print("+B")

    elif result>60:
        print("C")

    else:
        print("Fail")
outer()
```
