```python
from bs4 import BeautifulSoup
html="""
<html>
<head>
<title>Hello Everybody</title>
</head>
<body>
<h1>IMRD SHIRPUR</h1>
<p>MCA DEAPARMENT</p>
</body>
</html>"""

soup=BeautifulSoup(html,'html.parser')
title=soup.title.string
print("Title:",title)
h=soup.h1.string
print("Heading:",h)
para=soup.p.string
print("Paragraph:",para)
```

```python
import re
text="Hello Everyone Kaise Ho"
s=re.search(r'e',text)
if s:
    print(s.group())

else:
    print("Not Found")

f=re.findall('e',text)
print(f)

s=re.split('K',text)
print(s)

se=re.search(r'\s',text)
print(se.start())

m=re.match('H',text)
if m:
    print(m.group())
else:
    print("not found")
```

```python
import re
data="Kaise Hoooooo"
s=re.search(r'\s',data)
print("The position of white space:",s.start())

se=re.search('a',data)
print(se.group())

sp=re.split('e',data)
print(sp)

fi=re.findall('o',data)
print(fi)

m=re.match('K',data)
print(m.group())
```
