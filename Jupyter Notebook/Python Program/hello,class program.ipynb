```python
def show(a,b):
    c=a+b
    d=a*b
    e=a/b
    f=a-b
    g=a%b
    print("The addition of two variable is:",c)
    print("The multiplication of two variale is:",d)
    print("The division of two variable is:",e)
    print("The substraction of two variable is:",f)
    print("The modulus of two variable is:",g)

show(4,2)
```

```python
class demo:
    def __init__(self,name,age):
        self.name=name
        self.age=age

    def show(self):
        print(F"Name:{self.name},Age:{self.age}")

obj=demo("chetan",21)
obj.show()
```
