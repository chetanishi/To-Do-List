```python
with open('chetan.txt', 'w') as a:
    a.write("Hello, this is a file handling example in Python.\n")
    a.write("Python makes file handling easy!")


with open('chetan.txt', 'r') as a:
    content = a.read()  
    print(content)

with open('chetan.txt','a') as file:
    r="Hello Everyone Kaise HO"
    print(r)
```
