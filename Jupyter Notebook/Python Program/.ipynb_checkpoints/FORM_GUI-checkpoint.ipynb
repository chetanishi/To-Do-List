```python
from tkinter import *
from tkinter import messagebox
top=Tk()
top.title("Form Program")
top.geometry("500x500")

def fun():
    messagebox.showinfo("Hii","OK Button is clicked")
    
lbl=Label(top,text="Name" ,height=5,width=5)
lbl.grid(row=0,column=0)
entry=Entry(top,width="15")
entry.grid(row=0,column=1)

lbl1=Label(top,text="Password")
lbl1.grid(row=1,column=0)
entry1=Entry(top,width=15,show="*")
entry1.grid(row=1,column=1)

lbl3=Label(top,text="Select Color")
ch=Checkbutton(top,text="Red")
ch1=Checkbutton(top,text="Green")
ch2=Checkbutton(top,text="Black")
lbl3.grid(row=3,column=0)
ch.grid(row=4,column=0)
ch1.grid(row=5,column=0)
ch2.grid(row=6,column=0)

lbl4=Label(top,text="Select Gender")
gender=IntVar()
rb=Radiobutton(top,text="Male",variable="gender",value="1")
rb1=Radiobutton(top,text="Female",variable="gender",value="2")
lbl4.grid(row=7,column=0)
rb.grid(row=8,column=0)
rb1.grid(row=8,column=1)

btn=Button(top,text="OK",width="10",command=fun)
btn.place(x=50,y=350)

text=Text(top,height=5,width=10)
text.grid(padx=10)
text.insert("1.0","Type something else")

top.mainloop() 
```
