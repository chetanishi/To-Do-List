```python
def main_function(a,b):

    def add():
        return a+b
    def substract():
        return a-b
    def multiply():
        return a*b
    def div():
        return a/b if b!=0 else "Divide by 0 is not allowed"
    print("Addition:",add())
    print("Subtrsction:",substract())
    print("Multiplication:",multiply())
    print("Division:",div())

main_function(5,6)
    
```
